```python
import pandas as pd

popularidad_noticias = pd.read_csv('OnlineNewsPopularity.csv', sep = ",")
```

```python
popularidad_noticias.columns
```

```python
def dia_de_la_semana(fila):
    if fila[' weekday_is_monday'] == 1:
        return 'Lunes'
    elif fila[' weekday_is_tuesday'] == 1:
        return 'Martes'
    elif fila[' weekday_is_wednesday'] == 1:
        return 'Miércoles'
    elif fila[' weekday_is_thursday'] == 1:
        return 'Jueves'
    elif fila[' weekday_is_friday'] == 1:
        return 'Viernes'
    elif fila[' weekday_is_saturday'] == 1:
        return 'Sábado'
    elif fila[' weekday_is_sunday'] == 1:
        return 'Domingo'
```

```python
popularidad_noticias['dia_de_la_semana'] = popularidad_noticias.apply(lambda row: dia_de_la_semana(row), axis = 1)
```

```python
def tipo_articulo(fila):
    if fila[' data_channel_is_lifestyle'] == 1:
        return 'Estilo de vida'
    elif fila[' data_channel_is_entertainment'] == 1:
        return 'Entretenimiento'
    elif fila[' data_channel_is_bus'] == 1:
        return 'Negocios'
    elif fila[' data_channel_is_socmed'] == 1:
        return 'Redes sociales'
    elif fila[' data_channel_is_tech'] == 1:
        return 'Tecnología'
    elif fila[' data_channel_is_world'] == 1:
        return 'Mundo'
    else:
        return 'Otro'
```

```python
popularidad_noticias['tipo_noticia'] = popularidad_noticias.apply(lambda row: tipo_articulo(row), axis = 1)
```
